```python
import tensorflow as tf
import numpy as np
from PIL import Image
%matplotlib inline
from matplotlib.pyplot import imshow

from openslidertfpy import MicroPatchReader, is_mock
```

```python
with tf.Graph().as_default():
    coordinator = tf.train.Coordinator()
    reader = MicroPatchReader(
        "sample.svs", coordinator, image_width=500, image_height=500, verbose=False
    )
    images, locations, levels = reader.get_inputs()
    results = list()
    
    with tf.Session() as sess:
        locations2read = [(0, 0), (500*16, 0)]
        level2read = 2
        
        reader.start_thread([(l, level2read) for l in locations2read])
        
        while not coordinator.should_stop():
            imgs, ls, vs = sess.run([images, locations, levels])
            results.extend([
                (i, l, v) for i, l, v in zip(imgs, ls, vs)
                if not is_mock(l, v)
            ])
```

```python
left_img, left_location, left_level = results[0]
left_img = Image.fromarray(left_img.astype(np.uint8))
```

```python
imshow(left_img)
```

```python
print(left_location, left_level)
```

```python
right_img, right_location, right_level = results[1]
right_img = Image.fromarray(right_img.astype(np.uint8))
```

```python
imshow(right_img)
```

```python
print(right_location, right_level)
```
